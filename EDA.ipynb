```python
cols = ["Title","Genre", "Languages","MaturityRating","Country","Tags","IMDbscore","Release Date","Summary"]
df = dataset[cols].copy()
```

```python
df.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
df['Summary'] = df['Summary'].fillna('')
tfidf_matrix = tfidf.fit_transform(df['Summary'])
```

```python
tfidf_matrix.shape
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
similarity_mat = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
similarity_mat.shape
```

```python
indices = pd.Series(df.index, index=df['Title']).drop_duplicates()
```

```python
indices[:10]
```

```python
def get_recommendations(Title,similarity_mat):
    
    # Get the index of the movie that matches the title
    idx = indices[Title]
   

    # Get the pairwsie similarity scores of all movies with that movie
    scores = list(enumerate(similarity_mat[idx]))
    
    # Sort the movies based on the similarity scores
    scores = sorted(scores,key=lambda x: x[1],reverse=True)

    # Get the scores of the 10 most similar movies
    high_sim = scores[1:11]
   

    # Get the movie indices
    movie_indices = [i[0] for i in high_sim]

    # Return the top 10 most similar movies
    return df['Title'].iloc[movie_indices]
```

```python
get_recommendations('To Kill a Child',similarity_mat)
```

```python
#Hard Filtering 

#languages
#Country
#IMDb_bin 

#Rest columns (Title not included) is in similarity matrix 
```

```python
#Filter the dataset on the basis of languages, Country, Genre and IMDb Score

def filter(X,language,country,genre,score):
    X_new = pd.DataFrame()
    X = X.loc[(data['IMDbscore'] >= score)]
    for i in range(len(X)):
        if language in X.iloc[i]["Languages"]:
            if country in X.iloc[i]["Country"]:
                if genre in X.iloc[i]["Genre"]:
                    X_new = X_new.append(X.iloc[i])
    return(X_new)

#dataset = filter(data,"english","india","drama",7) 
```

```python
#Create the dataset to be fed to the recommendation system 
```

```python
#df_new = dataset.drop(["Country","Genre","IMDbscore","Languages","Release Date"],axis = 1) #already filtered
df_new = dataset.drop(["IMDbscore","Release Date"],axis = 1)
```

```python
df_new = df_new.reset_index(drop = True)
```

```python
#merge all white spaces in tags
def tags(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

df_new["Tags"] = df_new["Tags"].apply(tags)
```

```python
#merge all columns together to create the document matrix
```

```python
#df_new["merged"] = df_new["MaturityRating"] + df_new["Tags"]
df_new["merged"] = df_new["MaturityRating"] + df_new["Tags"] + df_new["Country"] + df_new["Genre"] + df_new["Languages"]
```

```python
df_new.head()
```

```python
df_new["Summary"]= df_new["Summary"].apply(lambda x: x.split(" "))
```

```python
df_new["body"] = df_new["Summary"]+df_new["merged"]
```

```python
#df_final = df_new.drop(["MaturityRating","Summary","Tags","merged"],axis = 1)
df_final = df_new.drop(["MaturityRating","Summary","Tags","merged","Country","Genre","Languages"],axis = 1)
```

```python
df_final.reset_index(drop = True)
```

```python
df_final["body"] = df_final['body'].apply(lambda x: ', '.join(map(str, x)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_final['body'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_mat = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df_final.index, index=df_final['Title']).drop_duplicates()
```

```python
def get_recommendations(Title,similarity_mat):
    
    # Get the index of the movie that matches the title
    idx = indices[Title]
   

    # Get the pairwsie similarity scores of all movies with that movie
    scores = list(enumerate(similarity_mat[idx]))
    
    # Sort the movies based on the similarity scores
    scores = sorted(scores,key=lambda x: x[1],reverse=True)

    # Get the scores of the 10 most similar movies
    high_sim = scores[1:11]
   

    # Get the movie indices
    movie_indices = [i[0] for i in high_sim]

    # Return the top 10 most similar movies
    return df_final['Title'].iloc[movie_indices]
get_recommendations('love',similarity_mat)
```
